```python
!pip install scikit-fuzzy
import skfuzzy as fuzz
import numpy as np
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
weight = ctrl.Antecedent(np.arange(0,101,1),'weight')
```

```python
weight['very_light']=fuzz.trimf(weight.universe,[0,0,35])
weight['light']=fuzz.trimf(weight.universe,[30,40,45])
weight['average']=fuzz.trimf(weight.universe,[45,50,65])
weight['heavy']=fuzz.trapmf(weight.universe,[50,65,100,100])
weight['very_heavy']=fuzz.trapmf(weight.universe,[75,75,100,100])

weight.view()
```
