```python
!pip install scikit-fuzzy
import skfuzzy as fuzz
import numpy as np
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
age = ctrl.Antecedent(np.arange(0,101,1),'age')
```

```python
age['veryYoung'] = fuzz.trimf(age.universe,[0,0,15])
age['young'] = fuzz.trapmf(age.universe,[0,12,20,25])
age['midAge'] = fuzz.trapmf(age.universe,[20,30,40,45])
age['old'] = fuzz.trapmf(age.universe,[40,50,70,70])
age['veryOld'] = fuzz.trapmf(age.universe,[50,70,100,100])

age.view()
```
